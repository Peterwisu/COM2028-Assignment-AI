```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Import the dataframe of training's result for each optimizers

```python
adam = pd.read_csv('logs/training_Adam.log')
adagrad = pd.read_csv('logs/training_Adagrad.log')
SGD = pd.read_csv('logs/training_SGD.log')
adadelta = pd.read_csv('logs/training_Adadelta.log')
adamax= pd.read_csv('logs/training_Adamax.log')
adam
```

## Select a style available from Matplotlib

```python
plt.style.available
```

```python
plt.style.use('seaborn-poster')
plt.style.use('seaborn-darkgrid')
#sns.set_theme()
```

## Plot a validation accuracy and loss between each optimizer

```python
fig , (ax0,ax1) = plt.subplots(nrows=1,ncols=2,figsize=(15,8))

# add data to ax0


ax0.plot(adam['val_accuracy'],label='Adam')
ax0.plot(adagrad['val_accuracy'],label='Adagrad')
ax0.plot(SGD['val_accuracy'],label='SGD')
ax0.plot(adamax['val_accuracy'],label='Adamax')
#ax0.plot(adadelta['val_accuracy'],label='adadelta')
# customize ax0 
ax0.set_yticks(np.arange(0.2,1
                         ,0.03))
ax0.set_title(label='Accuracy',fontsize=40)
ax0.set_xlabel(xlabel='Number of Epochs',fontsize=40)
ax0.set_ylabel(ylabel='val_accuracy',fontsize=40)
ax0.set_ylim([0.41,0.69])
ax0.legend(prop={"size":25});


ax1.plot(adam['val_loss'],label='Adam')
ax1.plot(adagrad['val_loss'],label='Adagrad')
ax1.plot(SGD['val_loss'],label='SGD')
ax1.plot(adamax['val_loss'],label='Adamax')
#ax1.plot(adadelta['val_loss'],label='adadelta')
# customize ax0 

ax1.set_title(label='Loss',fontsize=40)
ax1.set_xlabel(xlabel='Number of Epochs',fontsize=40)
ax1.set_ylabel(ylabel='val_loss',fontsize=40)
ax1.set_yticks(np.arange(.8,2.7,0.15))
ax1.legend(prop={"size":25});
fig.suptitle('Testing with different optimizer ',fontsize=50,fontweight='bold');
fig.tight_layout()
```

## Plot the learning curve of overffitng model

```python
overfitting= pd.read_csv('logs/training_Overfitting.log')
```

```python
overfitting.head()
```

```python
fig , ax = plt.subplots(figsize=(15,10))

ax.plot(overfitting['accuracy'],label='Training')
ax.plot(overfitting['val_accuracy'],label='Validation')

ax.set_xlabel(xlabel='Number of Epochs',fontsize=40)
ax.set_ylabel(ylabel='Accuracy',fontsize=40)
ax.set_ylabel('Accuracy')
ax.text(127, 0.75, '     Overfitting \n (High Variance)', fontsize = 28,fontweight='bold')
  
ax.errorbar(x=[125,125],y=[0.62,0.93],linestyle='--',color='r',marker='o')
ax.legend(prop={"size":25});
fig.suptitle('Overfitting Model',fontsize=50,fontweight='bold');
```
